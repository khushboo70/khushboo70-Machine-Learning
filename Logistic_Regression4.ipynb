```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('/content/drive/MyDrive/machine learning/Regression/dataset/penguins_binary_classification.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['species'].unique()
```

```python
df['island'].unique()
```

```python
df['island']!='Torgersen'
```

```python
df=df[df['island']!='Torgersen']
```

```python
df.head()
```

```python
df['species']=df['species'].map({'Adelie':0,'Gentoo':1})
```

```python
df['island']=df['island'].map({'Biscoe':0,'Dream':1})
```

```python
df.head()
```

```python
X=df.drop('island',axis=1)
y=df['island']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
```

```python
lg.fit(X_train,y_train)
```

```python
from sklearn import metrics
y_pred=lg.predict(X_test)
print("Accuracy of Logistic Regression is :",metrics.accuracy_score(y_pred,y_test))
```
